```python
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import img_to_array, load_img
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_path = '/Users/chenjialu/Desktop/DL_Assignment2/Assignment-2-Dataset-Round-1/data/train/1/img002-00174.png'
# img001-00001.png
```

```python
img = load_img(img_path)  # this is a PIL image
img = img_to_array(img)
plt.imshow(img)
img = np.expand_dims(img, axis=0)
```

### Example of image augmentation A:

```python
datagen = ImageDataGenerator(
                            zoom_range=0.2,
                            shear_range=0.2,
                            rescale=1. / 255,
                            rotation_range=30    
                           )
datagen.fit(img)
i=0
batch_size = 9
for img_batch in datagen.flow(img, batch_size=9):
    im = img_batch[0]
    plt.subplot(330 + 1 + i)
    plt.imshow(im)
    
    i=i+1
    if i >= batch_size:
        break
        
```

### Example of image augmentation  B:

```python

datagen = ImageDataGenerator(
                            zoom_range=0.1,
                            shear_range= 0.6,
                            rescale=1. / 255,
                            rotation_range= 6    
                           )
datagen.fit(img)
i=0
batch_size = 9
for img_batch in datagen.flow(img, batch_size=9):
    im = img_batch[0]
    plt.subplot(330 + 1 + i)
    plt.imshow(im)
    
    i=i+1
    if i >= batch_size:
        break
        
```
